```python
import numpy as np
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import joblib
import os

# 1. Load the preprocessed data we saved in Phase 3
X_train = np.load('../data/X_train.npy')
y_train = np.load('../data/y_train.npy')

# Safe print to avoid copy-paste formatting errors
print("Training models on", X_train.shape[0], "samples...\n")

# 2. Initialize the 5 approved algorithms
models = {
    "Logistic_Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Decision_Tree": DecisionTreeClassifier(random_state=42),
    "Random_Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(probability=True, random_state=42), 
    "Linear_Regression": LinearRegression()
}

# 3. Create a directory to save the trained models if it doesn't exist
os.makedirs('../models', exist_ok=True)

# 4. Train and save each model
for name, model in models.items():
    print("Training", name, "...")
    model.fit(X_train, y_train)
    
    # Save the trained model so we don't have to retrain every time
    model_path = '../models/' + name + '.pkl'
    joblib.dump(model, model_path)
    print("Model saved to", model_path, "\n")

print("All 5 models trained and saved successfully! Ready for Phase 5.")
```
