```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import joblib

# 1. Load the data
df = pd.read_csv('../data/alzheimers_prediction_dataset.csv')

# 2. Lock in ONLY the 13 features matching the React form
columns_to_keep = [
    "Age", "Education Level", "BMI", "Cognitive Test Score", 
    "Physical Activity Level", "Smoking Status", "Alcohol Consumption", 
    "Diabetes", "Hypertension", "Cholesterol Level", 
    "Family History of Alzheimer’s", "Depression Level", "Sleep Quality"
]
X = df[columns_to_keep]

# 3. Get the exact One-Hot Encoded columns
X_encoded = pd.get_dummies(X, drop_first=True)
joblib.dump(list(X_encoded.columns), '../models/expected_columns.pkl')
print("Saved expected column structure.")

# 4. Re-fit and save the Scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
joblib.dump(scaler, '../models/scaler.pkl')
print("Saved StandardScaler.")

# 5. Re-fit and save PCA (using dynamic 0.95 variance to perfectly match training)
pca = PCA(n_components=0.95, random_state=42)
pca.fit(X_scaled)
joblib.dump(pca, '../models/pca.pkl')
print("Saved PCA Transformer.")

print("\nPipeline artifacts successfully synced with the leaner model!")
```
